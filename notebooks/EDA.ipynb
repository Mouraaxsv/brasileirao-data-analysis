## Tratamento dos valores ausentes

- A base não apresentou linhas duplicadas.
- As estatísticas detalhadas das partidas apresentam grande quantidade de valores ausentes entre 2003 e 2017.
- Os valores ausentes das estatísticas não foram substituídos por zero, pois a ausência de informação não representa necessariamente a ocorrência de zero eventos.
- Foi identificada uma partida sem gols registrados nas colunas `gols_mandante` e `gols_visitante`.
- Como os gols são informações essenciais para as análises futuras, essa partida foi removida da base analítica.

```python
#Novas colunas de data

df_clean["ano"] = df_clean["data"].dt.year
df_clean["mes"] = df_clean["data"].dt.month
df_clean["dia"] = df_clean["data"].dt.day
```

```python
#Remover espaços em branco no início e no final das strings

colunas_texto = [
    "time_mandante",
    "time_visitante",
    "tecnico_mandante",
    "tecnico_visitante",
    "estadio",
    "arbitro"
]

for coluna in colunas_texto:
    df_clean[coluna] = df_clean[coluna].str.strip()
```

```python
# Coluna com o total de gols 

df_clean["total_gols"] = (
    df_clean["gols_mandante"] +
    df_clean["gols_visitante"]
)
```

```python
df_clean[
    ["gols_mandante",
     "gols_visitante",
     "total_gols"]
].head()
```

```python
def resultado(row):
    if row["gols_mandante"] > row["gols_visitante"]:
        return "Mandante"
    elif row["gols_mandante"] < row["gols_visitante"]:
        return "Visitante"
    else:
        return "Empate"

df_clean["resultado"] = df_clean.apply(resultado, axis=1)
```

```python
df_clean["resultado"].value_counts()
```

```python
# Salvar a base de dados limpa em um novo arquivo CSV

df_clean.to_csv(
    "../data/processed/brasileirao_clean.csv",
    index=False
)
```
